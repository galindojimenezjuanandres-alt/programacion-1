**Escritura de archivos en python**





**python permite cargar y crear objtos de archivos dados y tambien permite crear archivos y editar archivos dados**

```python
# crear y escribir un archivo de texto
with open("archivo.txt", 'w') as f:
     f.write("hola, este es un archivo de texto. \n")
     f.write("segunda linea. \n")
     f.write("esto es una prueba. \n")
```

**Con el argumento "w" se crea el archivo si no existe. tambien sobrescribe el archivo si ya existe, borrando su contenido previo**

Con archivos.txt
