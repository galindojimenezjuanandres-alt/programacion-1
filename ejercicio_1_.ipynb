```
def leer():
    while True:
        try:
            return float(input())
        except:
            print("Error: ingrese un numero valido.")

def leer_entero():
    while True:
        try:
            return int(input())
        except:
            print("Error: ingrese un numero entero valido.")

def suma(z1, z2):
    resultado = z1 + z2
    return resultado

def multiplicacion(z1, z2):
    resultado = z1 * z2
    return resultado

def raiz_n(z, n):
    resultado = z ** (1/n)
    return resultado

def pedir_complejo():
    print("Ingrese la parte real:")
    real = leer()
    print("Ingrese la parte imaginaria:")
    imag = leer()
    z = complex(real, imag)
    return z

opcion = 0

while opcion != 4:
    print("\nMENU DE OPCIONES")
    print("1. Sumar numeros complejos")
    print("2. Multiplicar numeros complejos")
    print("3. Raiz n-esima de un numero complejo")
    print("4. Salir")

    print("Seleccione una opcion:")
    opcion = leer_entero()

    if opcion == 1:
        z1 = pedir_complejo()
        z2 = pedir_complejo()
        r = suma(z1, z2)
        print("La suma es:", r)

    elif opcion == 2:
        z1 = pedir_complejo()
        z2 = pedir_complejo()
        r = multiplicacion(z1, z2)
        print("La multiplicacion es:", r)

    elif opcion == 3:
        z = pedir_complejo()
        print("Ingrese el valor de n:")
        n = leer_entero()

        if n == 0:
            print("Error: n no puede ser 0.")
        else:
            r = raiz_n(z, n)
            print("La raiz es:", r)

    elif opcion == 4:
        print("Saliendo del programa...")

    else:
        print("Opcion invalida, intente nuevamente.")
```
