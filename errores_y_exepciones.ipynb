```python
usuarios = {
    "ANA": "1234",
    "JUAN": "abcd"
}

try:
    usuario = input("Usuario: ").upper()
    password = input("Contraseña: ")

    if usuario not in usuarios:
        raise KeyError("El usuario no existe")

    if usuarios[usuario] != password:
        raise ValueError("Contraseña incorrecta")

    else:
        print("Bienvenido,", usuario)

except KeyError as e:
    print("Error:", e)

except ValueError as e:
    print("Error:", e)
```

```python
try:
    num1 = float(input("Ingrese el primer número: "))
    num2 = float(input("Ingrese el segundo número: "))
    op = input("Ingrese la operación (+, -, *, /, %): ")

    if op == "+":
        print("La suma es:", num1 + num2)

    elif op == "-":
        print("La resta es:", num1 - num2)

    elif op == "*":
        print("La multiplicación es:", num1 * num2)

    elif op == "/":
        if num2 == 0:
            raise ZeroDivisionError("No se puede dividir entre cero")
        print("La división es:", num1 / num2)

    elif op == "%":
        if num2 == 0:
            raise ZeroDivisionError("No se puede calcular módulo entre cero")
        print("El residuo es:", num1 % num2)

    else:
        raise ValueError("Operación no válida")

except Exception as e:
    print("Error:", e)
    exit()
```

```python
try:
    datos = input("Ingresa números separados por coma: ")
    lista = [float(x) for x in datos.split(",")]

    indice = int(input("Índice a consultar: "))
    print("Elemento:", lista[indice])

except ValueError:
    print("Debes ingresar solo números")

except IndexError:
    print("El índice está fuera de rango")
except TypeError:
      print("Los indices son numeros enteros mayores que 1")
```
