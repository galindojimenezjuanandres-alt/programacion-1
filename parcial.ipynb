```
def parcial(m):

if isinstance(m, int) and m >= 0:
  for n in range(0, (m+1)):
    if n == 1 or n == 0:
      print("El número es ", n)
    else:
      for x in range(2, n):
        if n % x == 0:
          print(n, 'es igual a', x, '*', n // x)
          break
      else: # Este else corresponde al for, no al if
       print(n, 'es un número primo')
else:
  print("El número no es natural")
```

```
palabra =input("increse una palabra ")

def contador_caracteres(palabra):
  letras = "abcdefghijklmnopqrstuvwxyz"
  letras1 = letras.upper()
  numeros = "o123456789"
  espacio = " "

  contador = 0
  contador1 = 0
  contador2 = 0
  contador3 = 0
  contador4 = 0

  for letras in palabra:
    if letras in letras:
      contador += 1
    elif letras in letras1:
      contador1 += 1
    elif letras in numeros:
      contador2 += 1
    elif letras in espacio:
      contador3 += 1
    else:
      contador4 += 1

    return({"cantidad de letras minusculas: ": contador,
            "cantidad de letras mayusculas :": contador1,
            "cantidad de numeros: ": contador2,
            "cantidad de espacios: ": contador3,
            "cantidad de caracteres especiales: ": contador4})
```

```
  contador_caracteres("holamundo123.%&/")
```

El codigo lo que haces es contar la cantidad de letras , mayusculas, numeros, espacios y caracteres especiales y sumarlos a un contador el cual vota como resultado la cantidad de minusculas, matusculas , numeros , espacios y caracteres que contiene la oracion o palabra
