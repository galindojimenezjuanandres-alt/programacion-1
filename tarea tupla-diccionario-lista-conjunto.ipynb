```
lista = [1, 2, 3, 4]

print("LISTA ORIGINAL:", lista)

# Lista a Tupla
tupla_desde_lista = tuple(lista)
print("Tupla:", tupla_desde_lista)

# Lista a Diccionario
diccionario_desde_lista = dict(enumerate(lista))
print("Diccionario:", diccionario_desde_lista)

# Lista a Conjunto
set_desde_lista = set(lista)
print("Conjunto:", set_desde_lista)

```

```
tupla = (2, 4, 8, 8)

print("TUPLA ORIGINAL:", tupla)

# Tupla a Lista
lista_desde_tupla = list(tupla)
print("Lista:", lista_desde_tupla)

# Tupla a Diccionario
diccionario_desde_tupla = dict(enumerate(tupla))
print("Diccionario:", diccionario_desde_tupla)

# Tupla a Conjunto
set_desde_tupla = set(tupla)
print("Conjunto:", set_desde_tupla)
```

```
diccionario = {"uno": 1, "dos": 2, "tres": 3}

print("DICCIONARIO ORIGINAL:", diccionario)

# Diccionario a Lista
lista_desde_dict = list(diccionario)
print("Lista:", lista_desde_dict)

# Diccionario a Tupla
tupla_desde_dict = tuple(diccionario.items())
print("Tupla:", tupla_desde_dict)

# Diccionario a Conjunto
set_desde_dict = set(diccionario)
print("Conjunto :", set_desde_dict)
```

```
conjunto = {100, 200, 300, 400}

print("CONJUNTO1:", conjunto)

# Conjunto a Lista
lista_desde_set = list(conjunto)
print("Lista:", lista_desde_set)

# Conjunto a Tupla
tupla_desde_set = tuple(conjunto)
print("Tupla:", tupla_desde_set)

# Conjunto a Diccionario
diccionario_desde_set = {x: x for x in conjunto}
print("Diccionario:", diccionario_desde_set)
```
